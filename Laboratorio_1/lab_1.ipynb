# Exploración de Datos
Lectura del conjunto de datos en formato .csv a *Python*, para posteriormente transformar la información en una estructura bidimensional: DataFrame (estructura de dato de la librería _Pandas_)

```python
# Importación de librería para el cálculo del coeficiente binomial
from math import comb
# Libreria Pandas
import pandas as pd
# Ajuste de Matplotlib
import matplotlib.pyplot as plt
plt.rcParams['svg.fonttype'] = 'none'

# Importanción del archivo .csv
musica = pd.read_csv('tcc_ceds_music.csv')

# 10 ejemplos del contenido del archivo .csv
musica.sample(10)
```

```python
# Conversión a un dataframe 
df_musica = pd.DataFrame(musica)
df_musica
```

```python
# Histograma de la columna 'loudness'
plt.hist(df_musica['loudness'], bins=50)

# Etiquetas de los ejes
plt.xlabel('Volumen')
plt.ylabel('Canciones')
```

```python
# Valores únicos de la columna 'genre'
generos = df_musica['genre'].nunique()
print("Los valores unicos son: ", generos)

# Cada valor unico cuantas veces aparece en la columna
cantidad_generos = df_musica['genre'].value_counts()
print("Las veces que aparecen los tipos de genre son: \n", cantidad_generos)
```

```python
# Valores únicos de la columna 'topic'
topic_unicos = df_musica['topic'].nunique()
print("Los valores unicos son: ", topic_unicos)

# Cada valor unico cuantas veces aparece en la columna
cantidad_topic = df_musica['topic'].value_counts()
print("Las veces que aparecen los tipos de topic son: \n", cantidad_topic)
```

```python
# Histograma de la columna 'danceability'
plt.hist(df_musica['danceability'], bins=50)

# Etiquetas de los ejes
plt.xlabel('Bailable')
plt.ylabel('Canciones')
```

```python
# Se asume que las canciones tienen una distribución normal en su caracteristica
# de que tan bailable es
# Cálculo de la media y desviación estandar
media = df_musica['danceability'].mean()

# Valores de la media  y desviación estandar
print(media)

# Definicion de la columna categorica en bailable o no, utilizando la media como umbral
bailable_o_no = []

for index, row in df_musica.iterrows():
    if (row['danceability'] <= media):
        bailable_o_no.append('No bailable')
    else:
        bailable_o_no.append('Bailable')

# Conversión de la columna "danceability" a variables categóricas
categoria_bailable_o_no = pd.DataFrame({'categoria_bailable_o_no': bailable_o_no})

df_musica = df_musica.join(categoria_bailable_o_no)

print(df_musica['categoria_bailable_o_no'])
```

```python
# Cantidad de ocurrencia de las categorias, "bailable" y "no bailable" de la columna "danceability"
nums_categoria_bailable_o_no = df_musica['categoria_bailable_o_no'].value_counts()
print(nums_categoria_bailable_o_no )
print(type(nums_categoria_bailable_o_no ))

# Cantidad de cada categoria
cantidad_bailable = nums_categoria_bailable_o_no.get('Bailable', 0)
cantidad_no_bailable = nums_categoria_bailable_o_no.get('No bailable', 0)
```

```python
# Graficación categórica de la columna "Bailable o no bailable"
plt.bar(nums_categoria_bailable_o_no.index, nums_categoria_bailable_o_no.values)
plt.xlabel("Bailables o no bailable")
plt.ylabel("Canciones")
plt.savefig('img/bailable.svg', transparent=True)
plt.show()
```

```python
# Conversión de colunma "loudness" a variables categóricas
valores_loud = []

for index, row in df_musica.iterrows():
    if (0 <= row["loudness"] <= 0.333):
        valores_loud.append("low")
    elif (0.333 < row["loudness"] < 0.666):
        valores_loud.append("mid")
    else:
        valores_loud.append("high")

categorical_loudness = pd.DataFrame({"categorical_loudness": valores_loud})

df_musica = df_musica.join(categorical_loudness)

print(df_musica["categorical_loudness"])
```

```python
# Cantidad de ocurrencia de las categorias, "low", "mid" y "high" de la columna "categorical_oudness"
nums_categorical_loudness = df_musica['categorical_loudness'].value_counts()
print(nums_categorical_loudness)
print(type(nums_categorical_loudness))
```

```python
# Graficación categórica de la columna "categorical_loudness"
plt.bar(nums_categorical_loudness.index, nums_categorical_loudness.values)
plt.xlabel("Magnitud de la onda sonora asociada a la canción del dataset")
plt.ylabel("Categoría de la magnitud de la onda sonora")
plt.savefig('img/intensidad_sonido.svg', transparent=True)
plt.show()
```

### Al reproducir al menos 2 canciones del _dataset_, ¿qué tan probable es que sean bailables?

### ¿Y 5 canciones del _dataset_?

Utilizando pruebas de Bernoulli se obtiene una probabilidad

```python
# Cantidad total de canciones categóricas (bailables y no bailables)
total_canciones = cantidad_bailable + cantidad_no_bailable

# Probabilidades de ser bailable o no bailable
p_bailables = cantidad_bailable / total_canciones
p_no_bailables = cantidad_no_bailable / total_canciones

# Muestra de 10 canciones
n = 10

# Elegir 2 canciones bailables
k_1 = 2
probabilidad_2_bailables = (comb(n, k_1)) * (p_bailables**k_1) * ((1 - p_bailables)**(n - k_1))

# Elegir 5 canciones bailables
k_2 = 5
probabilidad_5_bailables = (comb(n, k_2)) * (p_bailables**k_2) * ((1 - p_bailables)**(n - k_2))

print(f"Probabilidad de elegir 2 canciones bailables: {probabilidad_2_bailables:.2f}")
print(f"Probabilidad de elegir 5 canciones bailables: {probabilidad_5_bailables:.2f}")
```
